
**Step 3:- Downloading the data from yahoo finance for ADF testing**

````
workbook = xlsxwriter.Workbook('/usr/local/lib/python3.8/dist-packages/google/colab/workbook.xls')

#workbook = xlsxwriter.Workbook('workbook.xls')

# By default worksheet names in the spreadsheet will be
# Sheet1, Sheet2 etc., but we can also specify a name.
worksheet = workbook.add_worksheet("Z-Score")

today1 = dt.datetime.today().strftime('%Y,%m,%d')
# Start from the first cell. Rows and
# columns are zero indexed.
row = 0
col = 0
#############################################################
#OLD DATE
start = (dt.datetime.today() + relativedelta(months=-12)).date()

#TODAYS DATE
end = dt.datetime.today().date()
#############################################################
worksheet.write(row, col, str(start) )
worksheet.write(row, col + 1, str(end))
row = 1

worksheet.write(row, col, "STOCK1" )
worksheet.write(row, col + 1, "STOCK1")
worksheet.write(row, col + 2, "CORRALATION ")
worksheet.write(row, col + 3, "ADF")
row = 2
#start = dt.datetime(2021,6,1).date()
#end = dt.datetime(2021,8,28).date()

#############################################################
list =  [
        ["^NSEI","^CNXIT"],
        ["^NSEI","MCX.NS"],
        ["^NSEI","^NSEBANK"],
        ["INFY.NS","TCS.NS"],
        ["TECHM.NS","WIPRO.NS"],
        ["WIPRO.NS","HCLTECH.NS"],
        ["INFY.NS","WIPRO.NS"],
        ["INFY.NS","HCLTECH.NS"],
        ["HDFCBANK.NS","KOTAKBANK.NS"],
        ["ICICIBANK.NS","AXISBANK.NS"],
        ["KOTAKBANK.NS","ICICIBANK.NS"],
        ["KOTAKBANK.NS","AXISBANK.NS"],
        ["HDFCBANK.NS","HDFC.NS"],
        ["FEDERALBNK.NS","RBLBANK.NS"],
        ["ACC.NS","AMBUJACEM.NS"],
        ["TATASTEEL.NS","HINDALCO.NS"],
        ["JSWSTEEL.NS","HINDALCO.NS"],
        ["ONGC.NS","GAIL.NS"],
        ["ONGC.NS","NTPC.NS"],
        ["BAJFINANCE.NS","BAJAJFINSV.NS"],
        ["PFC.NS","RECLTD.NS"],
        ["AMARAJABAT.NS","EXIDEIND.NS"], 
        ["APOLLOTYRE.NS","MRF.NS"],
        ["ESCORTS.NS","BALKRISIND.NS"],
        ["DLF.NS","LT.NS"],
        ["DLF.NS","GODREJPROP.NS"],
        ["ULTRACEMCO.NS","GRASIM.NS"],
        ["ULTRACEMCO.NS","SHREECEM.NS"],
        ["tatamotors.NS","BAJAJ-AUTO.NS"],
        ["BAJAJ-AUTO.NS","heromotoco.NS"],
        ["TVSMOTOR.NS","heromotoco.NS"],
        ["DIVISLAB.NS","DRREDDY.NS"],
        ["DIVISLAB.NS","CIPLA.NS"],
        ["HINDPETRO.NS","IOC.NS"],
        ["HINDPETRO.NS","BPCL.NS"],
        ["SBILIFE.NS","HDFCLIFE.NS"],
        ["ASIANPAINT.NS","BERGEPAINT.NS"],
        ["SUNPHARMA.NS","AUROPHARMA.NS"],
        ["ICICIPRULI.NS","HDFCLIFE.NS"],
        ["JUBLFOOD.NS","TATACONSUM.NS"],
        ["NTPC.NS","POWERGRID.NS"],
        ["BRITANNIA.NS","HINDUNILVR.NS"],
        ["DABUR.NS","HINDUNILVR.NS"],
        ["WHIRLPOOL.NS","CROMPTON.NS"],
   #     ["MINDTREE.NS","LTI.NS"],
        ["HDFCBANK.NS","^NSEBANK"],
        ["SBIN.NS","KOTAKBANK.NS"],
        ["INDUSINDBK.NS","KOTAKBANK.NS"],
        ["CIPLA.NS","GLENMARK.NS"],
        ["ACC.NS","SHREECEM.NS"],
        ["PETRONET.NS","GAIL.NS"],
        ["BHEL.NS","LT.NS"],
        ["SUNPHARMA.NS","CIPLA.NS"],
        ["BIOCON.NS","DRREDDY.NS"],
        ]
  
# getting length of list
length = len(list)
   
# Iterating the index
# same as 'for i in range(len(list))'# write data to worksheet row by row and save the workbook after each row
for i in range(length):

    data1 = yf.download(list[i][0], start=start, end=end)
    data2 = yf.download(list[i][1], start=start, end=end)
    
    data = pd.concat([data1["Adj Close"], data2["Adj Close"]], axis=1)
    data.columns = [list[i][0], list[i][1]]
    
    # Calculate Z-Score
    data["Z-Score"] = (data.iloc[:,1] - data.iloc[:,1].rolling(window=20).mean()) / data.iloc[:,1].rolling(window=20).std()
    
    data.fillna(0, inplace=True)

    # Calculate Correlation
    corr = data.iloc[:,0].corr(data.iloc[:,1])

    # Perform ADF test
    adf = adfuller(data.iloc[:,1])

    # write the data to worksheet
    worksheet.write(row, 0, list[i][0])
    worksheet.write(row, 1, list[i][1])
    worksheet.write(row, 2, corr)
    worksheet.write(row, 3, adf[0])

    

    print(str(list[i][0])+"==="+list[i][1]+ "=== cor "+ str(corr) +"=== adf" + str(adf[0]))

    # save the workbook
    # move to the next row
    row += 1

# save the workbook when finished
workbook.close()
#/usr/local/lib/python3.7/dist-packages/xlsxwriter/workbook.py
# Open the Excel file in the default application for viewing Excel files
#os.startfile('/usr/local/lib/python3.7/dist-packages/xlsxwriter/workbook.xls')

files.download('/usr/local/lib/python3.8/dist-packages/google/colab/workbook.xls')
````
